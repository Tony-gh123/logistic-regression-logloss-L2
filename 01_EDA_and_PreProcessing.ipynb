```python
from datasets import load_dataset
import pandas as pd
import numpy as np
```

Data: https://huggingface.co/datasets/Cainiao-AI/LaDe



```python
ds = load_dataset("Cainiao-AI/LaDe", streaming=True)
print(ds)
```

```python
ds.keys()
```

```python
train = ds["train"]
sample_iter = train.shuffle(seed=42).take(10000)
df = pd.DataFrame(sample_iter)

df.head()
```

```python
df_columns = [
    'order_id',
    'from_dipan_id',
    'from_city_name',
    'delivery_user_id',
    'poi_lng',
    'poi_lat',
    'aoi_id',
    'typecode',
    'receipt_time',
    'receipt_lng',
    'receipt_lat',
    'sign_time',
    'ds'
]
```
